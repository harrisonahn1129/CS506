```python
class X:
    def __init__(self):
        pass

    def __repr__(self):
        return "X"

    def simplify(self):
        return self
    
    def evaluate(self, i):
        return Int(i)

class Int:
    def __init__(self, i):
        self.i = i
    
    def __repr__(self):
        return str(self.i)
    
    def simplify(self):
        return self
    
    def evaluate(self, i):
        return self

class Add:
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2
    
    def __repr__(self):
        
        # Cases of Add
        if isinstance(self.p1, Add):
            if isinstance(self.p2, Add):
                return "( " + repr(self.p1) + " ) + ( " + repr(self.p2) + " )"
            if isinstance(self.p2, Sub):
                return "( " + repr(self.p1) + " ) + ( " + repr(self.p2) + " )"
            return "( " + repr(self.p1) + " ) + " + repr(self.p2)
        if isinstance(self.p2, Add):
            return repr(self.p1) + " + ( " + repr(self.p2) + " )"
        
        # Cases of Sub
        if isinstance(self.p1, Sub):
            if isinstance(self.p2, Sub):
                return "( " + repr(self.p1) + " ) - ( " + repr(self.p2) + " )"
            if isinstance(self.p2, Add):
                return "( " + repr(self.p1) + " ) - ( " + repr(self.p2) + " )"
            return "( " + repr(self.p1) + " ) - " + repr(self.p2)
        if isinstance(self.p2, Sub):
            return repr(self.p1) + " - ( " + repr(self.p2) + " )"
        
        return repr(self.p1) + " + " + repr(self.p2)
    
    # Simplify Add
    def simplify(self):
        simp1 = self.p1.simplify()
        simp2 = self.p2.simplify()
        
        if isinstance(simp1, Int):
            # when only simp1 is 0
            if simp1.i == 0:
                return simp2.simplify()
            # when simp1 and simp2 are Int
            if isinstance(simp2, Int):
                return Int(simp1.i + simp2.i)
            # when both simp1 and simp2 are 0's
            if simp1.i == 0:
                if isinstance(simp2, Int):
                    if simp2.i == 0:
                        return Int(0)
            if isinstance(simp2, Add):
                sp1 = simp2.p1.simplify()
                sp2 = simp2.p2.simplify()
                if isinstance(sp1, Int):
                    if sp1.i == 0:
                        return Add( simp1, sp2).simplify()
                    if sp1.i == 0 and simp1.i == 0:
                        return sp2.simplify()
                    if simp1.i == 0:
                        return Add( sp1, sp2).simplify()
                    summ = simp1.i + sp1.i
                    return Add( Int(summ), sp2).simplify()
                if isinstance(sp2, Int):
                    if sp2.i == 0:
                        return Add( simp1, sp1).simplify()
                    if sp2.i == 0 and simp1.i == 0:
                        return sp1.simplify()
                    if simp1.i == 0:
                        return Add (sp1, sp2).simplify()
                    summ = simp1.i + sp2.i
                    return Add( Int(summ), sp1).simplify()

        # when simp2 is int
        if isinstance(simp2, Int):
            # when simp1 is func and simp2 is 0
            if simp2.i == 0:
                return simp1.simplify()            
            if isinstance(simp1, Add):
                sp1 = simp1.p1.simplify()
                sp2 = simp1.p2.simplify()
                if isinstance(sp1, Int):
                    if sp1.i == 0:
                        return Add( simp2, sp2).simplify()
                    if sp1.i == 0 and simp2.i == 0:
                        return sp2.simplify()
                    if simp2.i == 0:
                        return Add( sp1, sp2).simplify()
                    summ = simp2.i + sp1.i
                    return Add( Int(summ), sp2).simplify()
                if isinstance(sp2, Int):
                    if sp2.i == 0:
                        return Add( simp2, sp1).simplify()
                    if sp2.i == 0 and simp2.i == 0:
                        return sp1.simplify()
                    if simp2.i == 0:
                        return Add( sp1, sp2).simplify()
                    summ = simp2.i + sp2.i
                    return Add( Int(summ), sp1).simplify()

        return Add(simp1, simp2)
    
    def evaluate(self, i):
        simp1 = self.p1.simplify()
        simp2 = self.p2.simplify()
        return Add(simp1.evaluate(i), simp2.evaluate(i)).simplify()
    
# Sub Class for substraction    
class Sub:
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2
        
    def __repr__(self):
        
        # Cases of Add
        if isinstance(self.p1, Add):
            if isinstance(self.p2, Add):
                return "( " + repr(self.p1) + " ) - ( " + repr(self.p2) + " )"
            if isinstance(self.p2, Sub):
                return "( " + repr(self.p1) + " ) - ( " + repr(self.p2) + " )"
            return "( " + repr(self.p1) + " ) - " + repr(self.p2)
        if isinstance(self.p2, Add):
            return repr(self.p1) + " - ( " + repr(self.p2) + " )"
        
        # Cases of Sub
        if isinstance(self.p1, Sub):
            if isinstance(self.p2, Sub):
                return "( " + repr(self.p1) + " ) - ( " + repr(self.p2) + " )"
            if isinstance(self.p2, Add):
                return "( " + repr(self.p1) + " ) - ( " + repr(self.p2) + " )"
            return "( " + repr(self.p1) + " ) - " + repr(self.p2)
        if isinstance(self.p2, Sub):
            return repr(self.p1) + " - ( " + repr(self.p2) + " )"
            
            
        return repr(self.p1) + " - " + repr(self.p2)
    
    #Simplify Sub
    def simplify(self):
        simp1 = self.p1.simplify()
        simp2 = self.p2.simplify()
        
        if isinstance(simp1, Int):
            if simp1.i == 0:
                if isinstance(simp2, Int):
                    return Int( - simp2.i).simplify()
            if isinstance(simp2, Int):
                return Int(simp1.i - simp2.i)
            # when both simp1 and simp2 are 0's
            if simp1.i == 0:
                if isinstance(simp2, Int):
                    if simp2.i == 0:
                        return Int(0)
            if isinstance(simp2, Sub):
                sp1 = simp2.p1.simplify()
                sp2 = simp2.p2.simplify()
                if isinstance(sp1, Int):
                    if sp1.i == 0:
                        return Add( simp1, sp2).simplify()
                    if simp1.i == 0:
                        return Sub( Int( - sp1.i), sp2).simplify()
                    sub = simp1.i - sp1.i
                    if sub == 0:
                        return sp2.simplify()
                    return Sub( Int(sub), sp2).simplify()
                if isinstance(sp2, Int):
                    if sp2.i == 0:
                        return Sub( simp1, sp1).simplify()
                    if simp1.i == 0:
                        return Sub( sp1, sp2).simplify()
                    sub = simp1.i + sp2.i
                    if sub == 0:
                        return sp1.simplify()
                    return Sub( Int(sub), sp1).simplify()
                    
        if isinstance(simp2, Int):
            if simp2.i == 0:
                return simp1.simplify()
            if isinstance(simp1, Sub):
                sp1 = simp1.p1.simplify()
                sp2 = simp1.p2.simplify()
                if isinstance(sp1, Int):
                    if sp1.i == 0:
                        return Sub( simp2, sp2).simplify()
                    sub = sp1.i - simp2.i
                    if sub == 0:
                        return sp2.simplify()
                    return Sub( Int(sub), sp2).simplify()
                if isinstance(sp2, Int):
                    if sp2.i == 0:
                        return Sub( sp1, simp2).simplify()
                    if sp2.i == 0 and simp2.i == 0:
                        return sp1.simplify()
                    sub = - sp2.i - simp2.i
                    if sub == 0:
                        return sp1.simplify()
                    return Sub( sp1, Int(sub)).simplify()
            
        return Sub(simp1, simp2)
    
    def evaluate(self, i):
        simp1 = self.p1.simplify()
        simp2 = self.p2.simplify()
        return Sub(simp1.evaluate(i), simp2.evaluate(i)).simplify()
    

class Mul:
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2
    
    def __repr__(self):
        if isinstance(self.p1, Add):
            if isinstance(self.p2, Add):
                return "( " + repr(self.p1) + " ) * ( " + repr(self.p2) + " )"
            if isinstance(self.p2, Sub):
                return "( " + repr(self.p1) + " ) * ( " + repr(self.p2) + " )"
            return "( " + repr(self.p1) + " ) * " + repr(self.p2)
        if isinstance(self.p2, Add):
            return repr(self.p1) + " * ( " + repr(self.p2) + " )"
        
        # Modified to include the case of substraction
        if isinstance(self.p1, Sub):
            if isinstance(self.p2, Sub):
                return "( " + repr(self.p1) + " ) * ( " + repr(self.p2) + " )"
            if isinstance(self.p2, Add):
                return "( " + repr(self.p1) + " ) * ( " + repr(self.p2) + " )"
            return "( " + repr(self.p1) + " ) * " + repr(self.p2)
        if isinstance(self.p2, Sub):
            return repr(self.p1) + " * ( " + repr(self.p2) + " )"
        
        # Cases of Mul
        if isinstance(self.p1, Mul):
            if isinstance(self.p2, Mul):
                return "( " + repr(self.p1) + " ) * ( " + repr(self.p2) + " )"
            return "( " + repr(self.p1) + " ) * " + repr(self.p2)
        if isinstance(self.p2, Mul):
            return repr(self.p1) + " * ( " + repr(self.p2) + " )"
        
        # Cases of Div
        if isinstance(self.p1, Div):
            if isinstance(self.p2, Div):
                return "( " + repr(self.p1) + " ) * ( " + repr(self.p2) + " )"
            return "( " + repr(self.p1) + " ) * " + repr(self.p2)
        if isinstance(self.p2, Div):
            return repr(self.p1) + " * ( " + repr(self.p2) + " )"
        
        return repr(self.p1) + " * " + repr(self.p2)
    
    # Simplify Mul
    def simplify(self):
        simp1 = self.p1.simplify()
        simp2 = self.p2.simplify()
        
        if isinstance(simp1, Int):
            if simp1.i == 0:
                return Int(0)
            # if mul by 1
            if simp1.i == 1:
                return simp2.simplify()
            if isinstance(simp2, Int):
                return Int(simp1.i * simp2.i)
        if isinstance(simp2, Int):
            if simp2.i == 0:
                return Int(0)
            # mul by 1
            if simp2.i == 1:
                return simp1.simplify()
            
        if isinstance(simp1, Pow):
            sp1 = simp1.p1.simplify()
            sp2 = simp1.p2.simplify()
            if isinstance(sp2, Int):
                if isinstance(sp1, X):
                    if isinstance(simp2, X):
                        return Pow(sp1, Int(sp2.i + 1)).simplify()
                    if isinstance(simp2, Pow):
                        sp3 = simp2.p1.simplify()
                        sp4 = simp2.p2.simplify()
                        if isinstance(sp3, X):
                            if isinstance(sp4, Int):
                                return Pow(sp1, Int(sp2.i + sp4.i)).simplify()
        if isinstance(simp2, Pow):
            sp1 = simp2.p1.simplify()
            sp2 = simp2.p2.simplify()
            if isinstance(sp2, Int):
                if isinstance(sp1, X):
                    if isinstance(simp1, X):
                        return Pow(sp1, Int(sp2.i + 1)).simplify()
                    
        if isinstance(simp1, X):
            if isinstance(simp2, X):
                return Pow(simp1, simp2).simplify()            
            
        return Mul(simp1, simp2)
    
    def evaluate(self, i):
        simp1 = self.p1.simplify()
        simp2 = self.p2.simplify()
        return Mul(simp1.evaluate(i), simp2.evaluate(i)).simplify()

# Div Class for division
class Div:
    def __init__(self, p1, p2, message = "Cannot divide by 0"):
        self.p1 = p1
        self.p2 = p2
        self.message = message
    
    def __repr__(self):
        # Error when anything is divided by 0
        if self.p2 == Int(0):
            return self.message
        
        # Cases of addition
        if isinstance(self.p1, Add):
            if isinstance(self.p2, Add):
                return "( " + repr(self.p1) + " ) / ( " + repr(self.p2) + " )"
            if isinstance(self.p2, Sub):
                return "( " + repr(self.p1) + " ) / ( " + repr(self.p2) + " )"
            return "( " + repr(self.p1) + " ) / " + repr(self.p2)
        if isinstance(self.p2, Add):
            return repr(self.p1) + " / ( " + repr(self.p2) + " )"
        
        # Cases of substraction
        if isinstance(self.p1, Sub):
            if isinstance(self.p2, Sub):
                return "( " + repr(self.p1) + " ) / ( " + repr(self.p2) + " )"
            if isinstance(self.p2, Add):
                return "( " + repr(self.p1) + " ) / ( " + repr(self.p2) + " )"
            return "( " + repr(self.p1) + " ) / " + repr(self.p2)
        if isinstance(self.p2, Sub):
            return repr(self.p1) + " / ( " + repr(self.p2) + " )"
        
        # Cases of Mul
        if isinstance(self.p1, Mul):
            if isinstance(self.p2, Mul):
                return "( " + repr(self.p1) + " ) / ( " + repr(self.p2) + " )"
            return "( " + repr(self.p1) + " ) / " + repr(self.p2)
        if isinstance(self.p2, Mul):
            return repr(self.p1) + " / ( " + repr(self.p2) + " )"
        
        # Cases of Div
        if isinstance(self.p1, Div):
            if isinstance(self.p2, Div):
                return "( " + repr(self.p1) + " ) / ( " + repr(self.p2) + " )"
            return "( " + repr(self.p1) + " ) / " + repr(self.p2)
        if isinstance(self.p2, Div):
            return repr(self.p1) + " / ( " + repr(self.p2) + " )"
        
        # Cases of div
        return repr(self.p1) + " / " + repr(self.p2)
    
    # Simplify Div
    def simplify(self):
        simp1 = self.p1.simplify()
        simp2 = self.p2.simplify()
        
        # Case when simp2 is 0
        if isinstance(simp2, Int):
            if simp2.i == 0:
                return self.message
            if simp2.i == 1:
                return simp1.simplify()
        
        if isinstance(simp1, Int):
            if simp1.i == 0:
                return Int(0)
            if isinstance(simp2, Int):
                return Int(simp1.i / simp2.i)
            
        if isinstance(simp1, Pow):
            sp1 = simp1.p1.simplify()
            sp2 = simp1.p2.simplify()
            if isinstance(sp2, Int):
                if isinstance(sp1, X):
                    if isinstance(simp2, X):
                        return Pow(sp1, Int(sp2.i - 1)).simplify()
                    if isinstance(simp2, Pow):
                        sp3 = simp2.p1.simplify()
                        sp4 = simp2.p2.simplify()
                        if isinstance(sp4, Int):
                            if isinstance(sp3, X):
                                return Pow(sp1, Int(sp2.i - sp4.i)).simplify()
                            
        if isinstance(simp2, Pow):
            sp1 = simp2.p1.simplify()
            sp2 = simp2.p2.simplify()
            if isinstance(sp2, Int):
                if isinstance(sp1, X):
                    if isinstance(simp1, X):
                        return Pow(sp1, Int(1 - sp2.i)).simplify()
            
        return Div(simp1, simp2)
    
    def evaluate(self, i):
        simp1 = self.p1.simplify()
        simp2 = self.p2.simplify()
        return Div(simp1.evaluate(i), simp2.evaluate(i)).simplify()
    
# power class
class Pow:
    
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2
    
    def __repr__(self):
        
        if isinstance(self.p1, Add) or isinstance(self.p1, Sub) or isinstance(self.p1, Mul) or isinstance(self.p1, Div) or isinstance(self.p1, Pow):
            if isinstance(self.p2, Int) or isinstance(self.p2, X):
                return "( " + repr(self.p1) + " )^" + repr(self.p2)
        
        if isinstance(self.p2, Add) or isinstance(self.p2, Sub) or isinstance(self.p2, Mul) or isinstance(self.p2, Div) or isinstance(self.p2, Pow):
            if isinstance(self.p1, Int) or isinstance(self.p1, X):
                return repr(self.p1) + "^( " + repr(self.p2) + " )"
        
        if isinstance(self.p1, Add) or isinstance(self.p1, Sub) or isinstance(self.p1, Mul) or isinstance(self.p1, Div) or isinstance(self.p1, Pow):
            if isinstance(self.p2, Add) or isinstance(self.p2, Sub) or isinstance(self.p2, Mul) or isinstance(self.p2, Div) or isinstance(self.p2, Pow):
                return "( " + repr(self.p1) + " )^( " + repr(self.p2) + " )"
                
        return repr(self.p1) + "^" + repr(self.p2)
    
    def simplify(self):
        simp1 = self.p1.simplify()
        simp2 = self.p2.simplify()
        
        # Case when the power is 0
        if isinstance(simp2, Int):
            if simp2.i == 0:
                return Int(1)
            if simp2.i == 1:
                return simp1.simplify()
        
        # Case when the base is Int
        if isinstance(simp1, Int):
            # when base is 0
            if simp1.i == 0:
                return Int(0)
            # when base is 1
            if simp1.i == 1:
                return Int(1)
            # when simp2 is also Int
            if isinstance(simp2, Int):
                power = 1
                for i in range(simp2.i):
                    power = power * simp1.i
                return Int(power)
            
        # Case when simp1 is Pow
        if isinstance(simp1, Pow):
            sp1 = simp1.p1.simplify()
            sp2 = simp1.p2.simplify()
            if isinstance(sp2, Int):
                if isinstance(simp2, Int):
                    return Pow(sp1, Int(sp2.i * simp2.i)).simplify()
        
        return Pow(simp1, simp2)
    
    
    def evaluate(self, i):
        simp1 = self.p1.simplify()
        simp2 = self.p2.simplify()
        return Pow(simp1.evaluate(i), simp2.evaluate(i)).simplify()

test1 = Pow(Int(4), X())
print(test1)
print(test1.simplify())
print()
test2 = Mul(Pow(X(), Int(4)), X())
print(test2)
print(test2.simplify())
print()
test3 = Add(Mul(X(), X()), Int(0))
print(test3)
print(test3.simplify())
```
